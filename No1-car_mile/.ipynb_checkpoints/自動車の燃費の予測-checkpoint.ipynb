```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot') 

dataFrame= pd.read_table("train.tsv", encoding='utf-8')
dataFrame = dataFrame[dataFrame.horsepower != '?']
print(dataFrame)
```

```python
sns.pairplot(dataFrame, diag_kind='kde', height=5, markers="+", x_vars=["cylinders", "displacement", "mpg", "weight", "acceleration"], y_vars=["horsepower"])
```

```python
sns.jointplot(x="mpg", y="weight", data=dataFrame, kind="reg", color="purple", ratio=5, 
              space=0.2, dropna=True, xlim=None, ylim=None)
```

```python
sns.jointplot(x="mpg", y="displacement", data=dataFrame, kind="hex", color="purple", ratio=5, 
              space=0.2, dropna=True, xlim=None, ylim=None)
```

```python
sns.jointplot(x="mpg", y="cylinders", data=dataFrame, kind="reg", color="purple", ratio=5, 
              space=0.2, dropna=True, xlim=None, ylim=None)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
X = dataFrame.drop("mpg", 1)
X = X.drop("id", 1)
X = X.drop("car name", 1)
X = X.drop("origin", 1)
X = X.drop("model year", 1)
Y = dataFrame.mpg
```

```python
X
```

```python
linear_regression.fit(X,Y)
```

```python
print(linear_regression.coef_)
```

```python
linear_regression.intercept_
```

```python
import sklearn.model_selection
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y)
```

```python
testFrame= pd.read_table("test.tsv", encoding='utf-8')

exceptQuestionFrame = testFrame[testFrame.horsepower != '?'].copy()
exceptQuestionFrame['horsepower'] = exceptQuestionFrame['horsepower'].astype(np.float64)
meanFrame = exceptQuestionFrame.mean()

questionFrame = testFrame[testFrame.horsepower == '?'].copy()
questionFrame['horsepower'] = meanFrame['horsepower'].astype(np.float64)

resultFrame = pd.concat([exceptQuestionFrame, questionFrame])
```

```python
a = linear_regression.coef_
b = linear_regression.intercept_
# print(testFrame['horsepower']*a[2])
# print(float(raw_input(testFrame['horsepower'])*a[2])
resultFrame['mpg'] = resultFrame['cylinders']*a[0] + resultFrame['displacement']*a[1] + resultFrame['horsepower']*a[2] + resultFrame['weight']*a[3] + resultFrame['acceleration']*a[4] + b
resultFrame[['id', 'mpg']]
resultFrame[['id', 'mpg']].to_csv('./submit.csv', header=False, index=False)
```
